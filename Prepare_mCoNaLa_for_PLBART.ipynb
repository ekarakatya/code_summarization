```python
import string 
import json

def my_split(x):
  temp = x.split()
  result = []
  for tok in temp:
    accum = ''
    for t in tok: 
      if t not in string.punctuation:
        accum += t
      else:
        if len(accum) > 0:
          result.append(accum)
        accum = ''
        result.append(t)
    if len(accum) > 0:
      result.append(accum)
  return result
```

```python
import json

# presaved mCoNaLa version

dataset = []
with open('/content/drive/MyDrive/mCoNaLa/mconala_original.json', 'r', encoding='utf-8') as f:
  for line in f:
    ex = json.loads(line)
    dataset.append(ex)

```

```python
%cd /content/
with open('test.jsonl', 'w', encoding='utf-8') as f:
  for elem in dataset:
    text = elem['rewritten_intent'].replace('`', '')
    code = elem['snippet'].replace('\\n', '').replace('\\t', '\t')
    result = dict()
    result['code'] = code
    result['docstring_tokens'] = my_split(text)
    f.write(json.dumps(result)+'\n')
```

```python
%cd /content
!pip install evaluate
!pip install rouge_score
!pip install sacrebleu
import evaluate
import json
```

```python
references = []
with open('/content/drive/MyDrive/mCoNaLa/mconala_final', 'r', encoding='utf-8') as f:
  for line in f:
    references.append(line.strip().replace('`', '').lower())
```

```python
print(len(references))
```

```python
references
```

```python
results = '/content/drive/MyDrive/PLBART-main/scripts/code_to_text/base_python_ru_XX/output.hyp'
```

```python
predictions = []
with open(results, 'r', encoding='utf-8') as f:
  for line in f:
     pred = line.strip().lower()
     predictions.append(pred)
```

```python
print(len(predictions))
```

```python
predictions
```

```python
# BLEU (HF)

bleu = evaluate.load("bleu", max_order=4, smooth=True) #Нужно учитывать регистр букв!
results = bleu.compute(predictions=predictions, references=references)
print(results)
```

```python
# ROUGE-L
#https://huggingface.co/spaces/evaluate-metric/rouge
```

```python
rouge_l = evaluate.load('rouge')
results = rouge_l.compute(predictions=predictions,
                        references=references,
                        tokenizer=lambda x: x.split(), 
                        rouge_types=['rougeL'])
print(results)
```

```python
# chrF
#https://huggingface.co/spaces/evaluate-metric/chrf
```

```python
chrf = evaluate.load("chrf")
```

```python
results = chrf.compute(predictions=predictions,
                      references=references,
                      lowercase=True)
print(results)
```

```python
results = chrf.compute(predictions=predictions,
                      references=references,
                      char_order=1,
                      lowercase=True)
print(results)
```
